# stem plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,5,1,2,3,5]
plt.stem(x,y)
plt.show()
```

```python
# dotted line (linefmt=':')
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,5,1,2,3,5]
plt.stem(x,y,linefmt=':')
plt.show()
```

```python
# dotted line (linefmt=':') with markerfmt
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,5,1,2,3,5]
plt.stem(x,y,linefmt=':',markerfmt='r+')
plt.show()
```

```python
# dotted line (linefmt=':') with markerfmt
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,5,1,2,3,5]
plt.stem(x,y,linefmt=':',markerfmt='ro')
plt.show()
```

```python
# using bottom
# dotted line (linefmt=':')
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,5,1,2,3,5]
plt.stem(x,y,linefmt=':',bottom=0)
plt.show()
```

```python
# using basefmt parameter for colour
# dotted line (linefmt=':')
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,5,1,2,3,5]
plt.stem(x,y,linefmt=':', basefmt='black')
plt.show()
```

```python
# using label parameter
# dotted line (linefmt=':')
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,5,1,2,3,2]
plt.stem(x,y,linefmt=':',label='Python')
plt.legend(loc=1)
plt.show()
```

```python
# orientation parameter
# dotted line (linefmt=':')
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,5,1,2,3,5]
plt.stem(x,y,linefmt=':',orientation='horizontal')
plt.show()
```
