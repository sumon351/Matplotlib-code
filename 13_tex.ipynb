```python
# text function
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,1,4,3]
plt.plot(x,y,color='r')
plt.title('Text Used')
plt.xlabel('Day')
plt.ylabel('python growth')
plt.text(2,2,'text1',style='italic',fontsize=15,bbox={'facecolor':'red'})
plt.show()
```

```python
# annotate function
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,1,4,3,6]
plt.plot(x,y,color='r')
plt.title('Text Used')
plt.xlabel('Day')
plt.ylabel('python growth')
plt.text(2,2,'text1',style='italic',fontsize=15,bbox={'facecolor':'red'})
plt.annotate('text2',xy=(2,1),xytext=(4,3),arrowprops=dict(facecolor='black',shrink=100))
plt.show()
```
