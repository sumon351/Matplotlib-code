```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['c','c++','java','python']
plt.pie(x)
plt.show()
```

```python
# label parameter using
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['c','c++','java','python']
plt.pie(x,labels=y)
plt.show()
```

```python
# using explode parameter
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
# using explode parameter with color
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
colors=['b','g','m','y']
plt.pie(x,labels=y,explode=ex,colors=colors)
plt.show()
```

```python
# showing percentage
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['b','g','m','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%')
plt.show()
```

```python
# radius parameter with label distance
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['b','g','m','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',radius=1.5,labeldistance=0.8)
plt.show()
```

```python
# radius parameter with label distance with started angle parameter(startangle)
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['b','g','m','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',radius=1.5,labeldistance=0.8,startangle=90)
plt.show()
```

```python
# text properties (fontsize,counterclock)
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['b','g','m','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',radius=1.5,labeldistance=0.8,textprops={'fontsize':15},counterclock=False)
plt.show()
```

```python
# using title,legend
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=['b','g','m','y']
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%')
plt.title('Pie Plot')
plt.legend(loc=2)
plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
# multiple pie chart
import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,20,10]
y=['c','c++','java','python']


plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=1)

plt.show()
```

```python
# multiple pie chart
import matplotlib.pyplot as plt
x=[10,20,30,40]

y=['c','c++','java','python']


plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors='w')

plt.show()
```
