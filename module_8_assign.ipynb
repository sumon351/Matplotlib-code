**1. Instructions:**
* ** Use <code>plt.figure()</code> and create a figure object, say fig.**
* ** Add an axis to the figure canvas (`fig`), `[left, bottom, width, height] is [0,0,1,1]`. <i>Hint: <code>add_axes()</code><i>**
* ** Plot the data x,y on the axes**
* ** Set the labels x and y**
* ** Set the title**
* ** Create legends as shown in the plot**

Match the output below: 

```python
#Code here
fig=plt.figure(figsize=(4,3))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,'g-',label='y=x*2')
axes.set_xlabel('x_label')
axes.set_ylabel('y_label')
axes.set_title('plot_title')
plt.legend(loc=4)
plt.show()
```

**2. Instructions**<br>
* ** Create a figure object**
* ** Put two axes on the figure at [0,0,1,1] and [0.1,0.6,0.2,0.2]**
* ** Plot data (x,z) on main axes**
* ** Plot data (x,z) on the small axes and set its x limit b/w 1, 2, y limit 0, 5**
* ** Replicate the figure to axes, titles, legends etc**

```python
#Code here
fig1=plt.figure(figsize=(4,3))
axes_main=fig1.add_axes([0,0,1,1])
axes_insert=fig1.add_axes([0.1,0.6,0.2,0.2])
# main figure
axes_main.plot(x,z,'r-',label='y_main=x_main**2')
axes_main.legend(loc=4)


axes_main.set_title('Title_main')
axes_main.set_xlabel('x_main')
axes_main.set_ylabel('y_main')
# insert figure
axes_insert.plot(x,z)
axes_insert.set_xlabel('x_insert')
axes_insert.set_ylabel('y_insert')
axes_insert.set_title('insert')
axes_insert.set_xlim(1,2)
axes_insert.set_ylim(0,5)


```

**4. Replicate the output using <code>subplots()</code>**

```python
#Code here
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(3, 1, figsize=(4, 4))
x = np.linspace(0,1)

# Me plotting empty plots for demonstration
for ax in axs:
    ax.plot(x, np.zeros_like(x))
    ax.set_ylim(0, 1)

# Me adjusting layout
plt.tight_layout()
plt.show()
```

**5. plot `(x,y)`, `(x,z)` and `(x,y)` on the above empty canvas to replicate the plot given below.**

```python
#Code here
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10)
y = x*2
z = x**2


# Me creating a figure with 3 subplots
fig, axs = plt.subplots(3, 1, figsize=(4, 4))

# The first plot: (x, y) with blue dashed line
axs[0].plot(x, y, 'b--')  # This code will create blue dashed line


# The second plot: (x, z) with red solid line
axs[1].plot(x, z, 'r-')  # This code will create red solid line


# The third plot: (x, y) with green solid line
axs[2].plot(x, y, 'g-')  # This code will create green solid line


# layout adjusting
plt.tight_layout()
plt.show()
```

**6. Resize the above plots to `figsize = (3,4)`**

```python
#Code here
x = np.arange(0,10)
y = x*2
z = x**2


# Me creating a figure with 3 subplots
fig, axs = plt.subplots(3, 1, figsize=(3, 4))

# The first plot: (x, y) with blue dashed line
axs[0].plot(x, y, 'b--')  # This code will create blue dashed line


# The second plot: (x, z) with red solid line
axs[1].plot(y, z, 'r-')  # This code will create red solid line


# The third plot: (x, y) with green solid line
axs[2].plot(x, y, 'g-')  # This code will create green solid line


# layout adjusting
plt.tight_layout()
plt.show()
```
