# scatter plot

```python
# basic
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,6,3]
plt.scatter(day,no)
plt.show()
```

```python
# using title and color
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[3,4,5,1,6,7,8]
plt.scatter(day,no,color='r')
plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No', fontsize=15)
plt.show()
```

```python
# using different different colors
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[3,4,5,1,6,7,8]
colors=['r','g','b','m','y','r','y']
plt.scatter(day,no,c=colors)
plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No', fontsize=15)
plt.show()
```

```python
# using dotted size
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[3,4,5,1,6,7,8]
colors=['r','g','b','m','y','r','y']
sizes=[100,200,150,44,600,400,200]
plt.scatter(day,no,c=colors,s=sizes)
plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No', fontsize=15)
plt.show()
```

```python
# using transparency (alpha=0.5)
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[3,4,5,1,6,7,8]
colors=['r','g','b','m','y','r','y']
sizes=[100,200,150,44,600,400,200]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.5)
plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No', fontsize=15)
plt.show()
```

```python
# using marker parameter , like (marker='*')
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[3,4,5,1,6,7,8]
colors=['r','g','b','m','y','r','y']
sizes=[100,200,150,44,600,400,200]
plt.scatter(day,no,c=colors,s=sizes,marker='*')
plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No', fontsize=15)
plt.show()
```

```python
# changing marker edgecolor
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[3,4,5,1,6,7,8]
colors=['r','g','b','m','y','r','y']
sizes=[100,200,150,44,600,400,200]
plt.scatter(day,no,c=colors,s=sizes,marker='*',edgecolor='black')
plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No', fontsize=15)
plt.show()
```

# cmap(color maps)

```python
# using cmap with colorbar
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[3,4,5,1,6,7,8]
colors=[10,22,17,66,44,38,10]
sizes=[100,200,150,44,600,400,200]
plt.scatter(day,no,c=colors,s=sizes,cmap='viridis')
plt.colorbar()
plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No', fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[3,4,5,1,6,7,8]
colors=[10,22,17,66,44,38,10]
sizes=[100,200,150,44,600,400,200]
plt.scatter(day,no,c=colors,s=sizes,cmap='viridis')
t=plt.colorbar()
t.set_label('ColorBar',fontsize=15)
plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No', fontsize=15)
plt.show()
```

```python
# using multiple scatter 
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[3,4,5,1,6,7,8]
no2=[6,8,3,6,5,8,2]
colors=[10,22,17,66,44,38,10]
sizes=[100,200,150,44,600,400,200]
plt.scatter(day,no,c=colors,s=sizes,cmap='viridis')
plt.scatter(day,no2,c='black')
t=plt.colorbar()
t.set_label('ColorBar',fontsize=15)
plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No', fontsize=15)
plt.show()
```
